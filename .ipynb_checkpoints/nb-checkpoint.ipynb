```python
class cell(mesa.Agent):
    def __init__(self, unique_id, model):
        super().__init__(unique_id, model)
        self.wealth = 1

        
        
class matrix(mesa.Model):
    def __init__(self, p, s1, s2):
        self.grid = mesa.space.SingleGrid(s1, s2, True)
        self.schedule = mesa.time.RandomActivation(self)
        i=0
        for q in self.grid.get_neighborhood(p, moore=True, radius=2,
                                                  include_center=False):
            a = cell(i, self)
            i += 1
            self.schedule.add(a)
            self.grid.place_agent(a, q)


```
