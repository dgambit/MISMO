```python


params = {"side": [30],"n_towns":3, "mobility": [{"model" :"gravity", "metric":"relevance", "alpha":1.5, "beta":.5},
                                                 {"model" :"gravity", "metric":"relevance", "alpha":1.5, "beta":1},
                                                 {"model" :"gravity", "metric":"relevance", "alpha":1.5, "beta":1.5},
                                                 {"model" :"gravity", "metric":"relevance", "alpha":1.5, "beta":2},
                                                 {"model" :"gravity", "metric":"relevance", "alpha":0.5, "beta":.5},
                                                 {"model" :"gravity", "metric":"relevance", "alpha":0.5, "beta":1},
                                                 {"model" :"gravity", "metric":"relevance", "alpha":0.5, "beta":1.5},
                                                 {"model" :"gravity", "metric":"relevance", "alpha":0.5, "beta":2},
                                                {"model" :"gravity", "metric":"relevance", "alpha":1, "beta":.5},
                                                 {"model" :"gravity", "metric":"relevance", "alpha":1, "beta":1},
                                                 {"model" :"gravity", "metric":"relevance", "alpha":1, "beta":1.5},
                                                 {"model" :"gravity", "metric":"relevance", "alpha":1, "beta":2},               
                                                {"model" :"gravity", "metric":"relevance", "alpha":2, "beta":.5},
                                                 {"model" :"gravity", "metric":"relevance", "alpha":2, "beta":1},
                                                 {"model" :"gravity", "metric":"relevance", "alpha":2, "beta":1.5},
                                                 {"model" :"gravity", "metric":"relevance", "alpha":2, "beta":2}
                                     ]}

iterations = 10

results = mesa.batch_run(
    SchellingModel,
    parameters=params,
    iterations=iterations,
    max_steps=100,
    number_processes=1,
    data_collection_period=1,
    display_progress=True,
)

```

```python
import pandas as pd
```

```python
results = pd.DataFrame(results)
```

```python
results["model"]= results["mobility"].apply(lambda m: m["model"])
results["metric"]= results["mobility"].apply(lambda m: m["metric"])
results["alpha"]= results["mobility"].apply(lambda m: m["alpha"] if "alpha" in m else None)
results["beta"]= results["mobility"].apply(lambda m: m["beta"] if "beta" in m else None)
```
